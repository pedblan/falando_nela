# Coleta de apartes do Plenario da Camara

Notebook Colab para clonar o repositorio, instalar dependencias e executar o coletor metadata-only de apartes do Plenario da Camara salvando dados no Google Drive.

A primeira celula monta o Drive antes de qualquer codigo do projeto. Os dados completos ficam em `MyDrive/falando_nela/data`; o clone do repositorio fica em `/content/falando_nela` e pode ser recriado a cada sessao.

A coleta grava paginas oficiais em `raw/camara/plenario_apartes/metadata/` e nao cria corpus textual em `ano=YYYY/mes=MM/`.

Estrategia de janelas:

1. consulta anual de existencia;
2. se o ano tiver resultado, for inconclusivo ou falhar, consulta os trimestres daquele ano;
3. se o trimestre tiver resultado, for inconclusivo ou falhar, consulta os meses daquele trimestre.

Consultas anuais e trimestrais sao apenas probes de descoberta em `metadata/`. O `record_type` principal precisa ter janela de no maximo um mes, e a base de apartes segue metadata-only.


```python
import os
from pathlib import Path

DATA_ROOT = Path("/content/drive/MyDrive/falando_nela/data")
os.environ["FALANDO_NELA_DATA_ROOT"] = str(DATA_ROOT)

for name in ["raw", "checkpoints", "logs", "manifests", "processed"]:
    (DATA_ROOT / name).mkdir(parents=True, exist_ok=True)

print("FALANDO_NELA_DATA_ROOT=", os.environ["FALANDO_NELA_DATA_ROOT"])
```

```python
import os
import subprocess
from pathlib import Path

REPO_URL = "https://github.com/pedblan/falando_nela.git"
REPO_DIR = Path("/content/falando_nela")
REPO_REF = ""  # opcional: branch, tag ou commit. Deixe vazio para usar o default remoto.

if not REPO_DIR.exists():
    subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=True)
else:
    subprocess.run(["git", "-C", str(REPO_DIR), "fetch", "--all", "--tags", "--prune"], check=True)
    if not REPO_REF:
        subprocess.run(["git", "-C", str(REPO_DIR), "pull", "--ff-only"], check=True)

if REPO_REF:
    subprocess.run(["git", "-C", str(REPO_DIR), "checkout", REPO_REF], check=True)

subprocess.run(["git", "-C", str(REPO_DIR), "remote", "set-url", "origin", REPO_URL], check=True)
os.chdir(REPO_DIR)
print("Repo em:", Path.cwd())
subprocess.run(["git", "status", "--short"], check=True)
```

## Parametros

Use a validacao curta primeiro. Ela roda em modo producao para confirmar escrita no Drive, mas passa `--sample` e `--sample-limit` para limitar a execucao.

A validacao cobre um ano completo para exercitar a arvore `ano -> trimestre -> mes`. O recorte de producao continua amplo (`1900-01-01`) para maximizar cobertura historica; para analise substantiva, use `2010-01-01` em diante.

Se `processed/parlamentares/v1` existir no Drive, o coletor usa `parlamentares_periodos` para consultar apenas deputados com mandato vigente em cada ano. Para a coleta completa, rode antes o caderno de parlamentares ou o preparo de parlamentares do backfill historico geral.


```python
from datetime import datetime, timezone

DATA_INICIO_VALIDACAO = "2025-01-01"
DATA_FIM_VALIDACAO = "2025-12-31"
SAMPLE_LIMIT_VALIDACAO = 10

DATA_INICIO_PROD = "1900-01-01"
DATA_FIM_PROD = "2026-05-18"


PARLAMENTARES_PERIODOS_CANDIDATES = [
    DATA_ROOT / "processed" / "parlamentares" / "v1" / "parquet" / "parlamentares_periodos.parquet",
    DATA_ROOT / "processed" / "parlamentares" / "v1" / "parlamentares_periodos.jsonl",
]

RUN_STAMP = datetime.now(timezone.utc).strftime("%Y%m%dT%H%M%SZ")
RUN_ID_VALIDACAO = f"colab-camara-plenario-apartes-validacao-{RUN_STAMP}"
RUN_ID_PROD = "prod-camara-plenario-apartes-baseline"

print("RUN_ID_VALIDACAO=", RUN_ID_VALIDACAO)
print("RUN_ID_PROD=", RUN_ID_PROD, "# fixo para permitir retomada com --resume")
print("parlamentares_periodos disponivel=", any(path.exists() for path in PARLAMENTARES_PERIODOS_CANDIDATES))
```

```python
from collections import Counter
from datetime import date
import json
import subprocess
import sys
from pathlib import Path

MODULE = "coleta.camara.plenario_apartes.collect"
SOURCE = "camara"
DATASET = "plenario_apartes"
RECORD_TYPE = "sitaq_apartes_search_page"
YEAR_PROBE_RECORD_TYPE = "sitaq_apartes_year_probe"
QUARTER_PROBE_RECORD_TYPE = "sitaq_apartes_quarter_probe"
DISCOVERY_RECORD_TYPES = set(['deputados_apartes_metadata', 'deputados_apartes_atual_metadata'])
ALLOWED_RECORD_TYPES = {RECORD_TYPE, YEAR_PROBE_RECORD_TYPE, QUARTER_PROBE_RECORD_TYPE, *DISCOVERY_RECORD_TYPES}


def summarize_payload(payload):
    html = payload.get("html")
    extracted = payload.get("resultados") or payload.get("results") or payload.get("chaves_extraidas")
    if isinstance(extracted, list):
        total_extraidos = len(extracted)
    else:
        total_extraidos = None
    return {
        "aparteante_consultado": payload.get("aparteante_consultado"),
        "page_number": payload.get("page_number"),
        "probe_label": payload.get("probe_label"),
        "probe_status": payload.get("probe_status"),
        "html_bytes": len(html) if isinstance(html, str) else None,
        "resultados_extraidos": total_extraidos,
    }


def run_collector(run_id, data_inicio, data_fim, sample=False, sample_limit=None, resume=True, stream=True):
    cmd = [
        sys.executable,
        "-m",
        MODULE,
        "--mode",
        "prod",
        "--run-id",
        run_id,
        "--data-inicio",
        data_inicio,
        "--data-fim",
        data_fim,
    ]
    if resume:
        cmd.append("--resume")
    if sample:
        cmd.append("--sample")
    if sample_limit is not None:
        cmd.extend(["--sample-limit", str(sample_limit)])

    print("Rodando:", " ".join(cmd))
    output_lines = []
    if stream:
        process = subprocess.Popen(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True,
            bufsize=1,
        )
        assert process.stdout is not None
        for line in process.stdout:
            print(line, end="", flush=True)
            output_lines.append(line.rstrip("\n"))
        returncode = process.wait()
    else:
        result = subprocess.run(cmd, check=False, text=True, capture_output=True)
        returncode = result.returncode
        if result.stdout:
            print(result.stdout)
            output_lines.extend(result.stdout.splitlines())
        if result.stderr:
            print(result.stderr)
            output_lines.extend(result.stderr.splitlines())

    manifest_path = manifest_from_output(output_lines, run_id)
    if returncode != 0:
        print(f"Coletor terminou com returncode={returncode}; a celula continuara para permitir inspecao de logs/autosave.")
    return manifest_path


def manifest_from_output(output_lines, run_id):
    for line in reversed(output_lines):
        candidate = line.strip()
        if candidate.endswith(".json"):
            path = Path(candidate)
            if path.exists():
                return path

    manifest_path = DATA_ROOT / "manifests" / f"{run_id}.json"
    autosave_path = DATA_ROOT / "manifests" / f"{run_id}.autosave.json"
    if manifest_path.exists():
        return manifest_path
    if autosave_path.exists():
        return autosave_path
    return None


def show_manifest(manifest_path):
    if manifest_path is None:
        print("Manifest nao foi informado pelo coletor.")
        return None

    manifest_path = Path(manifest_path)
    if not manifest_path.exists():
        print("Manifest nao encontrado:", manifest_path)
        return None

    manifest = json.loads(manifest_path.read_text(encoding="utf-8"))
    print("\nManifest:", manifest_path)
    print(json.dumps(manifest, ensure_ascii=False, indent=2)[:6000])
    return manifest


def tail_log_for_run(run_id, n=30):
    log_path = DATA_ROOT / "logs" / f"{run_id}.jsonl"
    print("\nLog:", log_path)
    if not log_path.exists():
        print("Log ainda nao existe.")
        return
    lines = log_path.read_text(encoding="utf-8").splitlines()
    print("\n".join(lines[-n:]))


def load_metadata_records(path):
    path = Path(path)
    if not path.exists():
        return []
    records = []
    invalid_lines = []
    blank_lines = 0
    with path.open("r", encoding="utf-8") as handle:
        for line_number, line in enumerate(handle, start=1):
            stripped = line.strip()
            if not stripped:
                blank_lines += 1
                continue
            try:
                records.append(json.loads(stripped))
            except json.JSONDecodeError as exc:
                invalid_lines.append((line_number, str(exc), stripped[:160]))
    if blank_lines:
        print("linhas em branco ignoradas=", blank_lines)
    if invalid_lines:
        print("linhas JSON invalidas ignoradas=", len(invalid_lines))
        for line_number, error, prefix in invalid_lines[:5]:
            print({"linha": line_number, "erro": error, "prefixo": prefix})
    return records

def parse_iso_date(value):
    return date.fromisoformat(value)


def period_dates(record):
    periodo = record.get("periodo", {})
    return parse_iso_date(periodo["data_inicio"]), parse_iso_date(periodo["data_fim"])


def is_single_month(record):
    start, end = period_dates(record)
    return start.year == end.year and start.month == end.month


def print_record_type_summary(records):
    counts = Counter(record.get("record_type") for record in records)
    print("\nRecord types:")
    for record_type, total in sorted(counts.items()):
        print(f"  {record_type}: {total}")
    return counts


def inspect_metadata_jsonl(path, max_rows=8):
    path = Path(path)
    print("\nArquivo:", path)
    if not path.exists():
        print("Nao encontrado")
        return []

    records = load_metadata_records(path)
    print("linhas=", len(records), "bytes=", path.stat().st_size)
    print_record_type_summary(records)
    print("\nAmostra:")
    for record in records[:max_rows]:
        payload = record.get("payload", {})
        summary = {
            "record_type": record.get("record_type"),
            "partition": record.get("partition"),
            "periodo": record.get("periodo"),
            "source_id": record.get("source_id"),
        }
        if isinstance(payload, dict):
            summary.update(summarize_payload(payload))
        print(summary)
    return records


def assert_no_monthly_corpus_path(source, dataset):
    raw_root = DATA_ROOT / "raw" / source / dataset
    monthly_paths = sorted(raw_root.glob("ano=*/mes=*")) if raw_root.exists() else []
    print("\nParticoes de corpus encontradas:", monthly_paths)
    if monthly_paths:
        raise AssertionError("Base metadata-only nao deve criar ano=YYYY/mes=MM")


def assert_metadata_only_records(records):
    if not records:
        raise AssertionError("Nenhum registro metadata encontrado")
    bad_partitions = [record for record in records if record.get("partition") != "metadata"]
    if bad_partitions:
        raise AssertionError(f"Registros fora de metadata/: {len(bad_partitions)}")
    unexpected = sorted({record.get("record_type") for record in records} - ALLOWED_RECORD_TYPES)
    if unexpected:
        raise AssertionError(f"Record types inesperados: {unexpected}")


def assert_adaptive_window_contract(records):
    counts = Counter(record.get("record_type") for record in records)
    if counts[YEAR_PROBE_RECORD_TYPE] == 0:
        raise AssertionError("Validacao deve conter ao menos um preflight anual")

    monthly_records = [record for record in records if record.get("record_type") == RECORD_TYPE]
    for record in monthly_records:
        if not is_single_month(record):
            raise AssertionError(f"Record principal cobre mais de um mes: {record.get('source_id')} {record.get('periodo')}")

    if monthly_records and counts[QUARTER_PROBE_RECORD_TYPE] == 0:
        raise AssertionError("Registros mensais sem preflight trimestral")

    print("\nContrato de janelas OK:")
    print("  preflight anual:", counts[YEAR_PROBE_RECORD_TYPE])
    print("  preflight trimestral:", counts[QUARTER_PROBE_RECORD_TYPE])
    print("  registros mensais principais:", len(monthly_records))


def inspect_and_validate_metadata(path, max_rows=8):
    records = inspect_metadata_jsonl(path, max_rows=max_rows)
    assert_metadata_only_records(records)
    assert_no_monthly_corpus_path(SOURCE, DATASET)
    assert_adaptive_window_contract(records)
    return records
```

## Validacao curta

Esta celula roda `coleta.camara.plenario_apartes.collect` em `--mode prod`, com `--sample`, para confirmar que o Drive esta montado e que a base metadata-only grava apenas em `metadata/`.

A validacao tambem checa o contrato de janelas: deve haver preflight anual; se houver registros principais, eles precisam cobrir no maximo um mes e devem ter passado por preflight trimestral.


```python
manifest_validacao_path = run_collector(
    RUN_ID_VALIDACAO,
    DATA_INICIO_VALIDACAO,
    DATA_FIM_VALIDACAO,
    sample=True,
    sample_limit=SAMPLE_LIMIT_VALIDACAO,
    resume=True,
)
manifest_validacao = show_manifest(manifest_validacao_path)
```

```python
metadata_path = DATA_ROOT / "raw" / SOURCE / DATASET / "metadata" / f"{RUN_ID_VALIDACAO}.jsonl"

records_validacao = inspect_and_validate_metadata(metadata_path, max_rows=8)
tail_log_for_run(RUN_ID_VALIDACAO, n=20)
```

## Coleta completa

Esta celula roda o periodo completo em modo producao, com `run_id` fixo e `--resume`. Se o Colab cair ou a fonte falhar em alguma parte, rode a mesma celula novamente.

Mesmo na coleta completa, consultas anuais e trimestrais permanecem em `metadata/`; nenhum arquivo `ano=YYYY/mes=MM/` deve aparecer para apartes.


```python
EXECUTAR_COMPLETA = False
RUN_ID_COMPLETA = RUN_ID_PROD

if EXECUTAR_COMPLETA:
    print("Rodando coleta completa com run_id fixo:", RUN_ID_COMPLETA)
    manifest_completa_path = run_collector(
        RUN_ID_COMPLETA,
        DATA_INICIO_PROD,
        DATA_FIM_PROD,
        sample=False,
        sample_limit=None,
        resume=True,
        stream=True,
    )
    manifest_completa = show_manifest(manifest_completa_path)
    metadata_completa_path = DATA_ROOT / "raw" / SOURCE / DATASET / "metadata" / f"{RUN_ID_COMPLETA}.jsonl"
    if metadata_completa_path.exists():
        inspect_metadata_jsonl(metadata_completa_path, max_rows=8)
        assert_no_monthly_corpus_path(SOURCE, DATASET)
else:
    print("Altere EXECUTAR_COMPLETA para True quando quiser iniciar a coleta completa.")
```
